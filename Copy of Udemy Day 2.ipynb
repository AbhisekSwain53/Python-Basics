### To count how many years, weeks & days left in your life...............

####Method 1

```python
age = int(input("Please enter your current age:-"))     

years_remaining = 90 - age
weeks_remaining = years_remaining * 52
days_remaining = years_remaining * 365
months_remaining = years_remaining * 12
print("You have {0} weeks remaining, {1} days remaining & {2} months remaining.".format(weeks_remaining, days_remaining, months_remaining))
```

Method 2

```python
age = input("Please enter your current age:-")
age_as_int = int(age)
years_remaining = 90 - age_as_int
weeks_remaining = years_remaining * 52
days_remaining = years_remaining * 365
months_remaining = years_remaining * 12
message = f"You have {days_remaining} days, {weeks_remaining} weeks, {months_remaining} months left."
print(message)
```
